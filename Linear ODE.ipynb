```julia
using DynamicalSystems, Plots
```

```julia
@inline @inbounds function linear3d(u, p, t)
    A = p[1]; b = p[2]
    du1 = A[1,1]*u[1] + A[1,2]*u[2] + A[1,3]*u[3] + b
    du2 = A[2,1]*u[1] + A[2,2]*u[2] + A[2,3]*u[3] + b
    du3 = A[3,1]*u[1] + A[3,2]*u[2] + A[3,3]*u[3] + b
    return SVector{3}(du1, du2, du3)
end

A = [1 2 3; 4 5 6; 7 8 9]
b = 2
dsLinear1 = ContinuousDynamicalSystem(linear3d, rand(3), [A,b])

tr = trajectory(dsLinear1, 100)
plot(tr[:, 1], tr[:, 2],tr[:, 2], leg=false)
```

```julia
@inline @inbounds function linear3d(u, p, t)
    A = p[1]; b = p[2]
    n = size(A,1)
    x = []
    for i = 1:n
        x = [x; u[i]]
    end
    du = A*x + b*ones(n,1)
    return SVector{3}(du)
end

A = [1 2 3; 4 5 6; 7 8 9]
b = 2
dsLinear1 = ContinuousDynamicalSystem(linear3d, rand(3), [A,b])

tr = trajectory(dsLinear1, 100)
plot(tr[:, 1], tr[:, 2],tr[:, 2], leg=false)
```
